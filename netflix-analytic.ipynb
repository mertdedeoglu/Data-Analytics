# What kind of type is watching most on Netflix ? 
- I want to know what kinds are most wacthing on Netflix. That is my code for the purpose of do practice.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns="show_id") # It's not necessary
```

```python
df.isnull().sum() # There isn't null value at type.
```

```python

plt.figure(figsize=(10,10))
sns.histplot(x = df.type , edgecolor = ".1" , shrink = .7 , color="cyan")
plt.xlabel("Type")
plt.ylabel("Value")
plt.title("Most wacthing type of Netflix")
plt.savefig("graph.png")
plt.show()
```
